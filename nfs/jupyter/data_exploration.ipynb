```python
data = Path('../data/raw') / 'switchback_bw'
files = list(data.glob('*'))

w = [file for file in files if 'w.' in file.name]
a = [file for file in files if 'wa.' in file.name]
d = [file for file in files if 'wd.' in file.name]

smallest_class = min([len(w), len(a), len(d)])
[len(w), len(a), len(d)]
```

```python
moves_to_drop = a

number_to_drop = len(moves_to_drop) - smallest_class
number_to_drop 
```

```python
[item.unlink() for item in random.choices(moves_to_drop, k=number_to_drop)]
```
